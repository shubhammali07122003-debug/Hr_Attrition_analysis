# Attrition Distribution


```python
# eda
attrition_count = hr['attrition'].value_counts()

plt.figure()
plt.bar(attrition_count.index,attrition_count.values)
plt.title("Attrition Distribution")
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
# Department-wise Employee Count

# hr.head()

dept_counts = hr['department'].value_counts()

plt.figure()
plt.bar(dept_counts.index,dept_counts.values)
plt.title("Employees by Department")
plt.xlabel('Department')
plt.ylabel('count')
plt.xticks(rotation = 0)
plt.show()
```

```python
# Monthly Income Distribution (Histogram)
plt.figure()
plt.hist(hr['monthlyincome'],bins=20)
plt.title("Monthly Income Distribution")
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")
plt.show()
```

```python
# Age distribution

plt.figure()
plt.hist(hr['agegroup'],bins=15)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
gender_attr = pd.crosstab(hr['gender'],hr['attrition'])

gender_attr.plot(kind='bar')
plt.title("Attrition by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()                          
```

```python
salary_attr = pd.crosstab(hr['salaryslab'],hr['attrition'])

salary_attr.plot(kind='bar')
plt.title("Attrition by Salary Slab")
plt.xlabel("Salary Slab")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
job_attr = hr[hr['attrition'] =='Yes']['jobrole'].value_counts()

plt.figure()
plt.barh(job_attr.index, job_attr.values)
plt.title("Attrition by Job Role")
plt.xlabel("Count")
plt.ylabel("Job Role")
plt.show()
```

```python
working_attr = hr.groupby('totalworkingyears')['attrition'].apply(
    lambda x: (x=="Yes").sum()
)

plt.figure()
plt.plot(working_attr.index, working_attr.values)
plt.title("Attrition by Total Working Years")
plt.xlabel("Total Working Years")
plt.ylabel("Attrition Count")
plt.show()
```
